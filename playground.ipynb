```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
cppin = np.load("test.npy")
print(np.max(cppin[:,:,0]), np.min(cppin[:,:,0]))
print(np.max(cppin[:,:,1]), np.min(cppin[:,:,1]))
print(np.max(cppin[:,:,2]), np.min(cppin[:,:,2]))
print((cppin.astype('uint8') + 128)[:,:,0][:8,:8])
plt.imshow(cv2.cvtColor(cppin.astype('uint8') + 128, cv2.COLOR_YCrCb2RGB))
```

```python
cppout = np.load("test_out_dct.npy")
print(np.max(cppout[:,:,0]), np.min(cppout[:,:,0]))
print(np.max(cppout[:,:,1]), np.min(cppout[:,:,1]))
print(np.max(cppout[:,:,2]), np.min(cppout[:,:,2]))

plt.imshow(cv2.cvtColor(cppout.astype('uint8') + 128, cv2.COLOR_YCrCb2RGB))
(cppout.astype('uint8'))[:,:,0][:8,:8]
```
